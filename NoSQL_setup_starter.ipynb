```python
# Change the data type from String to Decimal for longitude and latitude

# Convert all string latitudes and longitudes to Decimal
update_latitude = establishments.update_many({"geocode.latitude": {"$type": "string"}}, {"$set": {"geocode.latitude": {"$toDouble": "$geocode.latitude"}}})
update_longitude = establishments.update_many({"geocode.longitude": {"$type": "string"}}, {"$set": {"geocode.longitude": {"$toDouble": "$geocode.longitude"}}})
print(f"Updated {update_latitude.modified_count} latitude values and {update_longitude.modified_count} longitude values.")
```

```python
# Convert RatingValueInt values to correct integer values
for doc in establishments.find({"RatingValueInt": {"$exists": True}}):
    rating_value = doc.get('RatingValueInt', None)
    if rating_value and isinstance(rating_value, str) and rating_value.isdigit():
        establishments.update_one(
            {"_id": doc["_id"]},
            {"$set": {"RatingValue": int(rating_value)}}
        )

# Convert latitudeDouble values to correct float values
for doc in establishments.find({"geocode.latitudeDouble": {"$exists": True}}):
    lat_value = doc.get('geocode', {}).get('latitudeDouble', None)
    if lat_value and isinstance(lat_value, str) and any(char.isdigit() for char in lat_value):
        establishments.update_one(
            {"_id": doc["_id"]},
            {"$set": {"geocode.latitude": float(lat_value)}}
        )

# Convert longitudeDouble values to correct float values
for doc in establishments.find({"geocode.longitudeDouble": {"$exists": True}}):
    long_value = doc.get('geocode', {}).get('longitudeDouble', None)
    if long_value and isinstance(long_value, str) and any(char.isdigit() for char in long_value):
        establishments.update_one(
            {"_id": doc["_id"]},
            {"$set": {"geocode.longitude": float(long_value)}}
        )

# Now, you can drop the renamed fields
establishments.update_many({}, {"$unset": {"RatingValueInt": "", "geocode.latitudeDouble": "", "geocode.longitudeDouble": ""}})

# Check the structure after updating
sample_document = establishments.find_one()
pprint(sample_document)
```
